```python
# Dependencies
import pandas as pd
```

```python
# Name of the CSV file
data = '../Resources/database.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(data, encoding="ISO-8859-1")
df
```

```python
# If wanting to delete columns
del df['Victim Count']
df.head()
```

```python
# If wanting to delete columns
del df['Perpetrator Count']
df.head()
```

```python
# Identify # of rows
df.count()
```

```python
# Drop all rows with missing information *turns out there aren't any*
df = df.dropna(how='any')
```

```python
 # Verify dropped rows
df.count()
```

```python
# What type of data is each column?
df.dtypes
```

```python
 # Use pd.to_numeric() method to convert the datatype of the Amount column
df['Perpetrator Age'] = pd.to_numeric(df['Perpetrator Age'])
```

```python
# Display an overview of the Weapon column
df['Weapon'].value_counts()
```

```python
# Display an overview of the Agency Type column
df['Agency Type'].value_counts()
```

```python
# Display a statistical overview
df['Victim Age'].describe().round()
```

```python
df.to_csv('../Resources/cleaned_database.csv', index=False)
```
